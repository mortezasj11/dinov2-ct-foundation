```python
import torch
from torchvision import transforms
from PIL import Image

# Check for GPU availability
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Using device: {device}")

# Load the DINOv2 model (adjust to match your model type, e.g., dinov2_vits14)
model = torch.hub.load('facebookresearch/dinov2', 'dinov2_vitb14')#dinov2_vitb14_reg
#model = torch.hub.load('facebookresearch/dinov2', 'dinov2_vitb14_reg')#dinov2_vitb14_reg
model = model.to(device)

# Load the backbone weights (without 'teacher' or other keys)
backbone_weights = torch.load('/rsrch1/ip/msalehjahromi/codes/dinov2-main/output_dir/eval/training_124999/teacher_checkpoint.pth', map_location=device)

if 'teacher' in backbone_weights:
    backbone_weights = backbone_weights['teacher']
```

```python
# Remove 'backbone.' prefix from the keys to match the model's state_dict format
processed_backbone_weights = {k.replace('backbone.', ''): v for k, v in backbone_weights.items()}

# Filter out keys that don't match the model's state_dict
model_state_dict = model.state_dict()
filtered_backbone_weights = {k: v for k, v in processed_backbone_weights.items() if k in model_state_dict}

# Load the filtered backbone weights into the model
model.load_state_dict(filtered_backbone_weights, strict=False)  # Use strict=False to ignore missing keys

# Set the model to evaluation mode
model.eval()

# # Load and preprocess the input image
# image_path = 'path/to/your/image.jpg'
# image = Image.open(image_path).convert('RGB')

# # Define image transformations (adjust as necessary)
# transform = transforms.Compose([
#     transforms.Resize((224, 224)),
#     transforms.ToTensor(),
#     transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
# ])

# # Apply transformations and move the image to the GPU
# image_tensor = transform(image).unsqueeze(0).to(device)  # Add batch dimension and move to GPU

# # Perform inference
# with torch.no_grad():
#     output = model(image_tensor)

# # For classification, get the predicted label
# _, predicted_class = output.max(1)
# print(f'Predicted class: {predicted_class.item()}')
```
